```python
# pace = 48 * ((Teams Possesion + Opponents Possesion) / (2 * (Teams Minutes Played / 5)))
# http://homepages.cae.wisc.edu/~ece539/fall13/project/Ferrero_rpt.pdf

# possession = 0.96*[(Field Goal Attempts)+(Turnovers)+0.44*(Free Throw Attempts)-(Offensive Rebounds)]
# https://www.nbastuffer.com/analytics101/possession/

# usage rate =
# 100*((Player’s Field Goal Attempts)+0.44*(Player’s Free Throw Attempts)+(Player’s Turnovers))*(Team’s Total Minutes)
# /
# ((Team’s Total Field Goal Attempts)+0.44*(Team’s Total Free Throw Attempts)+Team’s Total Turnovers))*5*(Player’s Minutes)
# https://www.nbastuffer.com/analytics101/usage-rate/

# defensive efficiency =
# Use the formula field goals attempted - offensive rebounds + turnovers + (0.4 x free throws attempted)
# = total number of possessions for the season to calculate total team possessions.
# Divide the total number of points allowed by your team by the possession total you calculated in Step 1. 
# For example, 1009 total points allowed divided by 998 total possessions gives your team 1.01 points allowed per possession.
# The opposite is true for defensive PPP: Above 1.0 is bad; below 1.0 is considered good.
# Convert defensive points per possession to a defensive efficiency rating by multiplying by 100.
# So 1.01 points allowed per possession becomes a defensive efficiency rating of 101.
# https://www.sportsrec.com/calculate-teams-offensive-defensive-efficiencies-7775395.html

def get_pace(team_pos, opp_pos, team_min):
    return 48 * ((team_pos + opp_pos) / (2 * (team_min / 5)))

def get_possession(fga, turnovers, fta, off_reb):
    return 0.96*((fga)+(turnovers)+0.44*(fta)-(off_reb))

def get_usage_rate(plr_fga, plr_fta, plr_turnover, plr_min, team_fga, team_fta, team_turnover, team_min):
    return 100*((plr_fga)+0.44*(plr_fta)+(plr_turnover))*(team_min) / ((team_fga)+0.44*(team_fta)+team_turnover)*5*(plr_min)
```
